```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('operation1.csv')

sns.lineplot(x=df.iloc[:,0] / 1000, y=df.iloc[:,1], color='red', label = "Experimento 1")
array = np.arange(1, 100, 1)
sns.lineplot(x=array, y=0.3*array*np.log(array), color='blue', linestyle='--', label="y=0.3 V log V").set_ylabel('Tempo (ms)')
sns.lineplot(x=array, y=0.1*array*np.log(array), color='yellow', linestyle='--', label="y=0.1 V log V").set_xlabel('Número de Vértices (x1000)')
```
